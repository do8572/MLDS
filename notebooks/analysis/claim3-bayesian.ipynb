```python
# standard data science toolbox
import numpy  as np
import pandas as pd

# calculate overtime for DT and 
from utils.helper_functions import dt_overtime, rf_overtime, rel_time, plot_rel_time

# experiment logs
HS_CLASS_CSV = "results/claim_1_1_dt_comparison_classification.csv"
CCP_CLASS_CSV = "results/claim_1_1_ccp_comparison_classification.csv"

HS_REGR_CSV = "results/claim_1_1_dt_comparison_regression.csv"
CCP_REGR_CSV = "results/claim_1_1_ccp_comparison_regression.csv"

RF_CSV = "results/rf_comparison_classification.csv"

# save results to 
DT_CLASS_TIME = "results/tables/dt-time-classification.tex"
DT_REGR_TIME = "results/tables/dt-time-regression.tex"
RF_TIME = "results/tables/rf-classification-time.tex"
```
